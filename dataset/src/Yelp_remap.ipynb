```python
import pandas as pd
import numpy as np
import random
```

```python
# set random seed
random.seed(42)
np.random.seed(42)
```

```python
datasets = ['Yelp']
```

```python
for dataset in datasets:
    print('Processing', dataset)
    # read data
    train_df = pd.read_csv(f'../processed/{dataset}/{dataset}.train.csv')
    valid_df = pd.read_csv(f'../processed/{dataset}/{dataset}.valid.csv')
    test_df = pd.read_csv(f'../processed/{dataset}/{dataset}.test.csv')

    asins = set(train_df['business_id'].unique())
    asins.update(train_df['history'].str.split().explode().unique())
   
    # remap asins
    asin_map = {asin: i for i, asin in enumerate(asins)}
    # new df with original asins, history and new remapped asins, history
    new_train_df = pd.DataFrame()
    new_valid_df = pd.DataFrame()
    new_test_df = pd.DataFrame()

    def remap_df(df):
        df['item_id'] = df['business_id'].map(asin_map)
        df['history_id'] = df['history'].str.split().apply(lambda x: ' '.join([str(asin_map[asin]) for asin in x]))
        return df

    new_train_df = remap_df(train_df)
    new_valid_df = remap_df(valid_df)
    new_test_df = remap_df(test_df)

    # save new data
    new_train_df.to_csv(f'../processed/{dataset}/{dataset}.train.remap.csv', index=False)
    new_valid_df.to_csv(f'../processed/{dataset}/{dataset}.valid.remap.csv', index=False)
    new_test_df.to_csv(f'../processed/{dataset}/{dataset}.test.remap.csv', index=False)
```
